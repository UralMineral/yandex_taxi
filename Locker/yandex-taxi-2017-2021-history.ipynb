```python
import numpy as
np
import pandas as pd
```

```python
df = pd.read_csv("orders.csv")
print('got', len(df.index), 'orders')
df
```

```python
df['created_at_utc'] = df['created_at_utc']
df['created_at_utc'] = pd.to_datetime(df.created_at_utc, unit='s')
df.sort_values(by=['created_at_utc'], inplace=True, ascending=False)
```

```python
df.head()
```

```python
df.tail()
```

```python
df = df.loc[df['status'] == 'finished']
print('got', len(df.index), 'finished orders')
```

```python
vehicle_model_df = df['vehicle_model'].value_counts()
print('got', len(vehicle_model_df.index), 'vehicle models')
ax = vehicle_model_df.head(25).plot.barh(x='val', y='lab', rot=0)
```

```python
vehicle_model_df.head(20).plot.pie(y='vehicle_model', figsize=(5, 5))
```

```python
df.head()
```

```python
df.loc[:, 'year'] = df['created_at_utc'].dt.year
df.loc[:, 'month'] = df['created_at_utc'].dt.month
```

```python
df
```

```python
# get orders by last month
last_month_df = df
last_month_df = last_month_df.loc[last_month_df.year.isin([2021]) & last_month_df.month.isin([5])]
print('got', len(last_month_df.index), 'orders at last month')
last_month_df.tail()
print('paid', np.floor(last_month_df['payment_cost'].sum()), 'RUR for last month with taxi')
```

```python
# get orders by last 3 month
months_df = df
months_df = months_df.loc[months_df.year.isin([2021]) & months_df.month.isin([3, 4, 5])]
print('got', len(months_df.index), 'orders at last 3 month')
months_df.tail()
print('paid', np.floor(months_df['payment_cost'].sum()), 'RUR for last 3 month with taxi')
```

```python
months_pie_df = pd.DataFrame(
{'month': [
    df.loc[df.year.isin([2021]) & df.month.isin([1])]['payment_cost'].sum(),
    df.loc[df.year.isin([2021]) & df.month.isin([2])]['payment_cost'].sum(),
    df.loc[df.year.isin([2021]) & df.month.isin([3])]['payment_cost'].sum(),
    df.loc[df.year.isin([2021]) & df.month.isin([4])]['payment_cost'].sum(),
    df.loc[df.year.isin([2021]) & df.month.isin([5])]['payment_cost'].sum(),
]},
index=['January', 'February', 'March', 'April', 'May'])
months_pie_df.head()
```

```python
months_pie_df.plot.pie(y='month', figsize=(5, 5))
print('total', np.floor(months_pie_df['month'].sum()), ' RUR')
```
